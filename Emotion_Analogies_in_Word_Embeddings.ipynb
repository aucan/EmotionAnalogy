```python
# Plutchik's Emotion Analogy
EmAnalogy=(
('optimism', 'anticipation', 'joy'), 
('hope', 'anticipation', 'trust'), 
('anxiety', 'anticipation', 'fear'), 
('love', 'joy', 'trust'), 
('guilt', 'joy', 'fear'), 
('delight', 'joy', 'surprise'), 
('submission', 'trust', 'fear'), 
('curiosity', 'trust', 'surprise'), 
('sentimentality', 'trust', 'sadness'), 
('awe', 'fear', 'surprise'), 
('despair', 'fear', 'sadness'),
('pride', 'anger', 'joy'),
('shame', 'fear', 'disgust'),
('dominance', 'anger', 'trust'))
```

```python
# Calculate Semantic Analogy on GloveTwitter Embeddings
for e1,e2,e3 in EmAnalogy:
    i=1
    for w,s in dict(model.most_similar(positive=[e2, e3], topn=20)).items():
        print(i,';',e1,';(',e2,'+',e3,');',w,';',model.similarity(e1,w))
        i=i+1
```
